```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Fidelity_plus = [0.3947, 0.3947, 0.4210, 0.4210, 0.3684, 0.0, -0.0526]
Fidelity_minus = [0.0526, 0.0526, 0.0526, 0.0789, 0.3947, 0.3684, 0.3684]
Sparsity = np.array([0.3530, 0.3530, 0.3566, 0.3703, 0.6688, 0.8929, 0.9152])
Sparsity_std = np.array([0.2224, 0.2224, 0.2234, 0.2234, 0.1612, 0.0741, 0.0471])
Contrastivity = np.array([0.8425, 0.8425, 0.8371, 0.8218, 0.5235, 0.2993, 0.2770])
Contrastivity_std = np.array([0.3048, 0.3048, 0.3062, 0.2999, 0.2438, 0.1568, 0.1495])
Threshold = np.array([0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 0.95])
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from cycler import cycler

colors = cycler(color=plt.get_cmap("tab10").colors)  # ["b", "r", "g"]
# colors = cycler(color=["#282782", "r", "g"])

mpl.style.use("ggplot")
mpl.rcParams["figure.figsize"] = (15, 5)
mpl.rcParams["axes.facecolor"] = "white"
mpl.rcParams["axes.grid"] = True
mpl.rcParams["grid.color"] = "lightgray"
mpl.rcParams["axes.prop_cycle"] = colors
mpl.rcParams["axes.linewidth"] = 1
mpl.rcParams["xtick.color"] = "black"
mpl.rcParams["ytick.color"] = "black"
mpl.rcParams["font.size"] = 12
mpl.rcParams["figure.titlesize"] = 25
mpl.rcParams["figure.dpi"] = 100
```

```python
# Plot the threshold and sparsity lists and show the plot
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(Threshold, Sparsity, label='Sparsity', marker='o')
ax[0, 0].fill_between(Threshold, Sparsity - Sparsity_std, Sparsity + Sparsity_std, color='b', alpha=0.2)
ax[0, 0].set_ylabel('Sparsity')
ax[0, 0].tick_params(axis='x', labelbottom=False)

ax[0, 1].plot(Threshold, Contrastivity, label='Contrastivity', marker='o')
ax[0, 1].fill_between(Threshold, Contrastivity - Contrastivity_std, Contrastivity + Contrastivity_std, color='b', alpha=0.2)
ax[0, 1].set_ylabel('Contrastivity')
ax[0, 1].tick_params(axis='x', labelbottom=False)

ax[1, 1].plot(Threshold, Fidelity_minus, label='Fidelity-', marker='o')
ax[1, 1].set_xlabel('Threshold')
ax[1, 1].set_ylabel('Fidelity-')
ax[1, 1].set_xlim(0.35, 1)

ax[1, 0].plot(Threshold, Fidelity_plus, label='Fidelity+', marker='o')
ax[1, 0].set_xlabel('Threshold')
ax[1, 0].set_ylabel('Fidelity+')
ax[1, 0].set_xlim(0.35, 1)

#plt.show()
plt.savefig('/Users/hecter/Desktop/thresholds.pdf', bbox_inches='tight')
```
