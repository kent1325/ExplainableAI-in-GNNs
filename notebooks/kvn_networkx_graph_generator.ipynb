```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
ax = plt.gca()

# Define the graph
G = nx.Graph()

# Add nodes to the graph
nodes = range(0, 12)
G.add_nodes_from(nodes)

# Define the edges with colors as well
# Assuming green ('C') nodes are 1, 3, 4, 6, 8 and yellow ('C') nodes are 2, 5, 7, 9, 10
edges = [(0, 1), (0, 2), (1, 3), (2, 4), (4, 5), (3, 5), (3, 6), (6, 7), (7, 8), (5, 8), (4, 9), (9, 10), (10, 11), (8, 11)]

# Add edges to the graph
G.add_edges_from(edges)

draw_args = dict()
draw_args["labels"] = {i: "C" for i in range(len(G.nodes))}
draw_args["node_color"] = [0 for n in G.nodes()]

# Draw the graph
pos = nx.kamada_kawai_layout(G)

# labels
ec = nx.draw_networkx_edges(G, pos, alpha=0.2)
lc = nx.draw_networkx_labels(
    G, pos, labels=draw_args["labels"], font_weight="bold", font_color="w"
)
nc = nx.draw_networkx_nodes(
    G,
    pos,
    ax=ax,
    cmap=plt.cm.viridis,
    node_size=500,
    node_color=draw_args["node_color"],
)

plt.axis('off')  # turn off the axis
#plt.savefig("frequent_substructure_positive.png", bbox_inches='tight')
plt.show()  # display
```
